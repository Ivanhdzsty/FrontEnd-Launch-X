## Caso Abogabot

**Descripcion general**

El despacho de abogados "Abogabot",  requiere automatizar las demandas de su cliente, a traves de la pagina web de su empresa que mediante el llenado de un formulario lo mandara al proceso de pago para finalizar la transaccion de la solictud, todo esto mediante una cuenta personal del cliente que tendra que crear inicialmente y donde podra ver el seguimiento de su solicitud del procedimiento legal.

Se tendra la cuenta de administrador del sitio web mediante la cual la cuenta  recibira las notificaciones de las demandas solicitadas y con la informacion del llenado del formulario que realizo el cliente, por lo cual se creara automaticamente el documento legal en formato word para empezar el proceso, el administrador podra ver si el pago fue realizado y visualizar en su dashboard para ver la cantidad de ingresos recibidos. Al mismo tiempo el administrador podra actualizar el proceso de la demanda y agregar comentarios en cada paso del proceso.

Adicional a esto la pagina debera ser responsive y la preferencia de colores del cliente (de la peticion) es azul marino y blanco.


**Requerimiento general:**

Automatizar las demandas del cliente.

**Requerimientos particulares** (donde se menciona la necesidad)
 
- Es un despacho de abogados que quiere automatizar las demandas de sus clientes, esto lo harán a traves de una página web llenando un formulario.

- El administrador del sitio recbe la notificación de una nueva demanda y con los datos llenados del formulario se crea automaticamente el documento legal en formato word para empezar el proceso.

- El administrador recibe el pago y debe de ser capaz de verlo en un dashboard para ver la cantidad de ingresos recibidos.

### Requerimientos funcionales

**Empresa desarroladora de la Aplicacion.**

- Toma de requerimiento.
- Diseño.
- Desarrollo.
- Verificacion.
- Mantenimiento y evaluacion por contrato.

**Usuario-Cliente Abogabot.**

- Inicio de Sesion.
- Registro.
- Informacion de la pagina principal (Del grupo Abogabot).
- Llenado de Formulario de la solicitud del proceso legal.
- Estado del pago de la solicitud (Verificado - No realizado).
- Seguimiento del proceso en su panel personal de solicitudes.
- Notificación por correo del avance de su proceso legal.

**Administrador Abogabot.**

- Recepcion de solicitud e informacion del solicitante.
- Dashboard de ingresos monetario.
- Control de solicitudes.
- Seguimiento y actualización de cada caso.
